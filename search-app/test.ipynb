```python
import json
import requests

f = open("API_keys.json")
data = json.load(f)

API_KEY = data["google"]["api_key"]
SEARCH_ENGINE_ID = data["google"]["search_engine_id"]
SERP_endpoint = data["google"]["SERP_endpoint"]
f.close()

print(API_KEY, SEARCH_ENGINE_ID, SERP_endpoint)

query="global climate change"

payload = {
        'key': API_KEY,
        'q': query,
        'cx': SEARCH_ENGINE_ID,
        'lr': "lang_en"
    }

SERP_response = requests.get(url=SERP_endpoint, params=payload)

search_results = SERP_response.json()
print(search_results)
```

```python
import json
import requests

f = open("API_keys.json")
data = json.load(f)

API_KEY = data["serp_api"]["api_key"]
SERP_endpoint = data["serp_api"]["SERP_endpoint"]
f.close()

print(API_KEY, SERP_endpoint)

query="global climate change"

payload = {
        'api_key': API_KEY,
        'engine': "google",
        'q': query
    }

SERP_response = requests.get(url=SERP_endpoint, params=payload)

search_results = SERP_response.json()
print(search_results["organic_results"])
```
